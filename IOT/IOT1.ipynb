```python
import numpy as np
import pandas as pd
import sklearn
import xml.etree.ElementTree as ET
import matplotlib.pyplot as plt
import os
import re
from lxml import etree
```

```python
X, Y = [], []
for i in os.listdir('Files'):
    try:
        with open('Files/'+i, 'r', encoding='utf-8') as f:
            xml = f.read()
        location = re.search('<location>(.*)</location>', xml).group(1)
        loc = location.split(',')
        X.append(int(loc[0]))
        Y.append(int(loc[1]))
    except:
        print(i)
        break
```

```python
plt.figure(figsize=(25, 25))
plt.scatter(X, Y)
```
